# Визуализация Plotly

```python
import pandas as pd
import numpy as np
import plotly
import cufflinks

from plotly.offline import iplot
import plotly.graph_objs as go
import plotly.figure_factory as ff
plotly.offline.init_notebook_mode()
cufflinks.go_offline()
```

```python
df_vacancy = pd.read_csv('hh_vacancy.csv')
df_vacancy = df_vacancy.dropna()
df_vacancy = df_vacancy.sample(frac=0.2)
df_vacancy.head()
```

### Ящик с усами

```python
factor = 'experience_name'
df_vacancy.pivot(columns=f'{factor}', values='salary_from').iplot(
    kind='box', yTitle='salary_from', title=f'salary_from distribution by {factor}')
```

## Распределение

```python
factor = 'experience_name'
hist_data, group_labels = [], []
for factor_value in df_vacancy[factor].unique():
    factor_value_data = list(df_vacancy.loc[df_vacancy[factor] == factor_value]['salary_from'])
    if len(factor_value_data) > 1:
        group_labels.append(factor_value)
        hist_data.append(factor_value_data)

iplot(ff.create_distplot(hist_data, group_labels, bin_size=10000))
```

## Гистограммы

```python
factor = 'experience_name'
df_vacancy.pivot(columns=f'{factor}', values='salary_from').iplot(
    kind='histogram',  bins=10, yTitle='salary_from', title=f'salary_from distribution by {factor}')
```

```python
df_vacancy.groupby(f'{factor}').size().iplot(kind='bar')
```

## Диаграмма рассеивания

```python
factor = 'experience_name'
df_vacancy.iplot(kind="scatter", x='salary_to', y="salary_from", categories=factor)
```
